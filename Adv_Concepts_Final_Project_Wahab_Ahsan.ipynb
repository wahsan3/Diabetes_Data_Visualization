Dataset has to equal 0 because the project is still able to run when I put a spelling error in the file path.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, shapiro, linregress

file_path = 'C:/Users/wahsa/Downloads/DiabetesAtlasData (1).csv'

#if the file is present, store it in the dataset variable, which will then be used in further operations.
try:
    dataset = pd.read_csv(file_path)
    
#else tell the user that the file is not present and subsequent operations will not run.
except FileNotFoundError:
    print(f"File not found: {file_path}")
    dataset = None
    
#if the dataset is present, print the first 5 rows of the file.  
if dataset is not None:
    print(f"First 5 rows of each column sorted by Diagnosed Diabetes-2021-Percentage:")
    print(dataset.head())
    
    print("\n\nSummary statistics for each column with numeric values:")
#Provides the quartiles, mean, and standard deviation.
    print(dataset.describe())
    
    print("\n\n")

    #To show that dataset is clean
    print(f"Sum of null data values in each column:\n{dataset.isnull().sum()}")

    #Define columns to keep using list comprehension
    columns_to_keep = [col for col in dataset.columns if col not in ["County", "State"]]

    #Create a new DataFrame with only the specified columns
    dataset2 = dataset[columns_to_keep]

    #Plotting the distribution of Diagnosed Diabetes Percentage
    plt.figure(figsize=(10, 6))
    sns.histplot(dataset2["Diagnosed Diabetes-2021-Percentage"], bins=20, kde=True, color="skyblue")
    plt.title("Distribution of Diagnosed Diabetes Percentage in 2021")
    plt.xlabel("Diagnosed Diabetes Percentage")
    plt.ylabel("Frequency")
    plt.show()
    #Calculate skewness and kurtosis using the column of interest.
    skewness = skew(dataset2["Diagnosed Diabetes-2021-Percentage"])
    kurt = kurtosis(dataset2["Diagnosed Diabetes-2021-Percentage"])

    print(f"Skewness: {skewness:.2f}")
    print(f"Kurtosis: {kurt:.2f}")

    #Calculate the t-statistic and p-value
    stat, p_value = shapiro(dataset2["Diagnosed Diabetes-2021-Percentage"])
    print(f"Shapiro-Wilk Test p-value: {p_value}")
    #alpha value is set to 0.05
    if p_value < 0.05:
        print("The data is likely not normally distributed.")
    else:
        print("The data is likely normally distributed.")

    # Scatter plot of Diagnosed Diabetes Percentage vs Overall SVI 2018 Percentile
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x="Diagnosed Diabetes-2021-Percentage", y="Overall SVI-2018-Percentile", data=dataset2, color="coral")
    sns.regplot(x="Diagnosed Diabetes-2021-Percentage", y="Overall SVI-2018-Percentile", data=dataset2, scatter=False, color="blue")
    plt.title("Diagnosed Diabetes Percentage vs Overall SVI 2018 Percentile")
    plt.xlabel("Diagnosed Diabetes Percentage")
    plt.ylabel("Overall SVI 2018 Percentile")
    plt.show()
    #Calculate correlation using base function method .corr()
    correlation_coefficient = dataset2["Diagnosed Diabetes-2021-Percentage"].corr(dataset2["Overall SVI-2018-Percentile"])
    print(f"Correlation Coefficient (r): {correlation_coefficient:.2f}")

    #https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html#scipy.stats.linregress
    slope, intercept, r_value, p_value, std_err = linregress(dataset["Diagnosed Diabetes-2021-Percentage"], dataset2["Overall SVI-2018-Percentile"])
    print(f"Slope: {slope:.2f}")
    print(f"y-intercept: {intercept:.2f}")
    print(f"R-squared: {r_value**2:.2f}")
    print(f"p_value: {p_value: .4f}")
    print(f"std_err: {std_err: .4f}")
    print(f"Equation of best-fit line: y = {slope:.2f}x {intercept:.2f}")

    #Bar chart of Top 10 Counties with the highest Diagnosed Diabetes Percentage
    top10_counties = dataset.nlargest(10, "Diagnosed Diabetes-2021-Percentage")
    plt.figure(figsize=(10, 6))
    sns.barplot(x="Diagnosed Diabetes-2021-Percentage", y="County", data=top10_counties, palette="viridis")
    plt.title("Top 10 Counties with the Highest Diagnosed Diabetes Percentage in 2021")
    plt.xlabel("Diagnosed Diabetes Percentage")
    plt.ylabel("County")
    plt.show()

    mean_diabetes_percentage = top10_counties["Diagnosed Diabetes-2021-Percentage"].mean()
    std_diabetes_percentage = top10_counties["Diagnosed Diabetes-2021-Percentage"].std()

    print(f"\nAdditional Statistics for Top 10 Counties:")
    print(f"Mean Diagnosed Diabetes Percentage: {mean_diabetes_percentage:.2f}")
    print(f"Standard Deviation Diagnosed Diabetes Percentage: {std_diabetes_percentage:.2f}")
    
```

The following is an analysis on the 3 data visualizations included above.

For the graph titled, "Distribution of Diagnosed Diabetes Percentage in 2021," the skewness and kurtosis values are positive, which indicate that the graph is positively skewed (right-skewed) and leptokurtic (high peak). The mean therefore is higher than the median in this distribution. Despite having a bell-shaped curve, the skewness is too right-sided, and thus, the p-value indicates that the sample does not come from a normally-distributed population. 

For the graph titled, "Diagnosed Diabetes Percentage vs Overall SVI 2018 Percentile," the correlation coefficient is 0.35, which has a small positive value, indicating that there is some positive increase in Diagnosed Diabetes Percentage as overall SVI 2018 percentile increases. For one unit increase in Diagnosed Diabetes Percentage, the Overall SVI 2018 Percentile increases 0.14 units. This is a poor linear regression model because only 12% of the variability in Overall SVI 2018 Percentile is explained by Diagnosed Diabetes Percentage. The standard error associated with the slope is 0.0387, which indicates a precise slope estimate.

For the graph titled, "Top 10 Counties with the Highest Diagnosed Diabetes Percentage in 2021," we see that the mean for diagnosed diabetes percentage is 9.10 and the standard deviation of diagnosed diabetes percentage is 0.92. Within 1 standard deviation of the mean, there is a distance of 0.92 units.
